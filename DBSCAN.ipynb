```python
#Mahlika George - Capstone Project

#importing the required packages
import pandas as pd
from pylab import rcParams
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import math as m

import sklearn
from sklearn.cluster import DBSCAN
from collections import Counter
```

```python
#initilizaig the grid size and style for matplotlib
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')

filename = '2007-2016.xlsx'
```

```python
from Cleaning.cluster import creating_clusters
Values = creating_clusters(filename)
```

```python
x = Values['x']
y = Values['y']
colors = Values['colors']

#setting up the subplot for years 2007-2010
fig, axs = plt.subplots(2, 2, sharex=True, sharey = True)
axs[0, 0].scatter(x[0], y[0], c = colors[0])
axs[0, 0].set_title('2007')
axs[0, 1].scatter(x[1], y[1], c = colors[1])
axs[0, 1].set_title('2008')
axs[1, 0].scatter(x[2], y[2], c = colors[2])
axs[1, 0].set_title('2009')
axs[1, 1].scatter(x[3], y[3], c = colors[3])
axs[1, 1].set_title('2010')

for ax in axs.flat:
    ax.set(xlabel='Population', ylabel='Clearance Average')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
```
